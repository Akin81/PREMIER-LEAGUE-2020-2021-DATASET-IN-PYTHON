```python
#WE NOW LOAD OUR DATASET
premier_df=pd.read_csv("C:\\Users\\User\\Desktop\\dataset\\EPL_20_21.csv")
premier_df.head()
```

```python
#TO GET THE NULL VALUES
premier_df.isna().sum()
```

```python
#FOR TOTAL NUMBER OF DUPLICATED VALUES
premier_df.duplicated().sum()
```

```python
# WE CREATE TWO ADDITIONAL COLUMN
premier_df ['MinsPerMatch'] = (premier_df['Mins']/ premier_df['Matches']).astype(int)
premier_df['GoalsPerMatch']= (premier_df['Goals']/premier_df['Matches']).astype(float)
premier_df
```

```python
#TOTAL GOALS SCORED
Total_Goals = premier_df['Goals'].sum()
print(Total_Goals)
```

```python
#TOTAL PENALTY GOALS
Total_PenaltyGoals= premier_df['Penalty_Goals'].sum()
print(Total_PenaltyGoals) 
```

```python
#PENALTY ATTEMPT
Total_PenaltyAttempt= premier_df['Penalty_Attempted'].sum()
print(Total_PenaltyAttempt)
```

```python
#TOTAL PENALTY MISSED
Penalty_NotScored= Total_PenaltyAttempt-Total_PenaltyGoals
print(Penalty_NotScored)
```

```python
#PIE CHART FOR PENALTY MISSED VS SCORES
plt.figure(figsize=(13,6))
Penalty_NotScore= Total_PenaltyAttempt-Total_PenaltyGoals
data= [Penalty_NotScore,Total_PenaltyGoals]
labels= ['Penalty_Missed','Penalty_Scored']
colors= sns.color_palette("Paired")
plt.pie (data,labels=labels,colors= colors,autopct ='%0f%%')
plt.show()
```

```python
#UNIQUE POSITIONS IN THE TEAM
premier_df ['Position'].unique()
```

```python
#NUMBER OF FORWARD PLAYERS
premier_df[premier_df['Position']=='FW']
```

```python
#PLAYERS FROM DIFFERENT NATION
np.size((premier_df['Nationality'].unique()))
```

```python
#MOST PLAYERS FROM WHICH COUNTRIES
nationality= premier_df.groupby('Nationality').size().sort_values(ascending= False)
nationality.head(10).plot(kind= 'bar',figsize=(12,6),color= sns.color_palette("mako"))
```

```python
#CLUBS WITH THE MAXIMUM NUMBER OF PLAYERS
premier_df['Club'].value_counts().nlargest(5).plot(kind= 'bar',color= sns.color_palette("flare"))
```

```python
#CLUBS WITH THE MAXIMUM NUMBER OF PLAYERS
premier_df['Club'].value_counts().nsmallest(5).plot(kind= 'bar',color= sns.color_palette("flare"))
```

```python
#PLAYERS BASED ON AGE GROUP
Under_20= premier_df[premier_df['Age']<= 20]
age_20_25=premier_df[(premier_df['Age']> 20) & (premier_df['Age']<= 25)]
age_25_30=premier_df[(premier_df['Age']> 25) & (premier_df['Age']<=30)]   
Above_30= premier_df[premier_df['Age']> 30]                    
```

```python
x = np.array([Under_20['Name'].count(),age_20_25 ['Name'].count(),age_25_30 ['Name'].count(),Above_30 ['Name'].count()])
my_labels= ['<=20','>20 & <=25','>25 & <=30','>30']
plt.title('Total Players with Age Groups', fontsize= 20)
plt.pie(x, labels=my_labels, autopct ="%.1f%%")
plt.show
```

```python
#NUMEBR OF UNDER 20 PLAYERS IN EACH CLUB
players_under_20= premier_df[premier_df['Age']< 20]
players_under_20 ['Club'].value_counts().plot(kind='bar',color= sns.color_palette("rocket_r"))
```

```python
#PLAYERS UNDER 20 PLAYING FOR MANU
players_under_20[players_under_20["Club"]== "Manchester United"]
```

```python
#PLAYERS UNDER 20 AT CHELSEA
players_under_20[players_under_20["Club"]== "Chelsea"]
```

```python
#AVERAGE AGE OF PLAYERS INE EACH CLUB
plt.figure(figsize= (12,6))
sns.boxplot(x= 'Club',y= 'Age', data = premier_df)
plt.xticks(rotation= 90)
```

```python
#CLUB AVERAGE AGE OF PLAYERS
num_player= premier_df.groupby('Club').size()
data=(premier_df.groupby('Club')['Age'].sum())/num_player
data.sort_values(ascending= False)
```

```python
#TOTAL ASSISTS FROM EACH CLUB
Assists_by_clubs=pd.DataFrame(premier_df.groupby('Club',as_index=False)['Assists'].sum())
sns.set_theme(style='whitegrid',color_codes= True)
ax=sns.barplot(x='Club', y= 'Assists',data=Assists_by_clubs.sort_values(by='Assists'),palette="rocket_r")
ax.set_xlabel('Club', fontsize= 30)
ax.set_ylabel('Assists', fontsize= 20)
plt.xticks(rotation =90)
plt.rcParams['figure.figsize']= (20,8)
plt.title('Plot of Clubs vs Total Assits' ,fontsize= 20)
```

```python
#TOP 10 ASSISTS
Top_10_Assists= premier_df[['Name','Club','Assists', 'Matches']].nlargest(n=10,columns='Assists')
Top_10_Assists
```

```python
#TOTAL GOALS FROM EACH CLUB
Assists_by_clubs=pd.DataFrame(premier_df.groupby('Club',as_index=False)['Goals'].sum())
sns.set_theme(style='whitegrid',color_codes= True)
ax=sns.barplot(x='Club', y= 'Goals',data=Assists_by_clubs.sort_values(by='Goals'),palette="rocket_r")
ax.set_xlabel('Club', fontsize= 30)
ax.set_ylabel('Goals', fontsize= 20)
plt.xticks(rotation =90)
plt.rcParams['figure.figsize']= (20,8)
plt.title('Plot of Clubs vs Total Goals' ,fontsize= 20)
```

```python
#MOST GOALS BY PLAYERS
Top_10_Goals= premier_df[['Name','Club','Goals', 'Matches']].nlargest(n=10, columns= 'Goals')
Top_10_Goals
```

```python
#GOALS PER MATCH
Top_10_Goals_per_Match= premier_df[['Name','GoalsPerMatch', 'Goals','Matches']].nlargest(n=10, columns= 'GoalsPerMatch')
Top_10_Goals_per_Match
```

```python
#PIE CHART GOALS WITH ASSISTS AND WITHOUT ASSISTS
plt.figure(figsize = (14,7))
assists= premier_df['Assists'].sum()
data= [Total_Goals - assists, assists]
labels=['Goals without assists','Goals with assists']
color= sns.color_palette("viridis")
plt.pie(data,labels=labels, colors =color, autopct= '%.0f%%')
plt.show()
```

```python
#TOP 10 PLAYERS WITH MOST YELLOW CARDS
premier_yellow= premier_df.sort_values(by='Yellow_Cards', ascending= False)[:10]
plt.figure(figsize = (20,6))
plt.title('Players with the most yellow cards')
c= sns.barplot(x= premier_yellow['Name'],y= premier_yellow['Yellow_Cards'],label= 'Players',color= 'yellow')
plt.ylabel= ('Number of Yellow cards')
c.set_xticklabels(c.get_xticklabels(),rotation=45)

```
